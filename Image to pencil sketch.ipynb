### Importing Library

```python
import cv2
```

### Importing Dataset

```python
image = cv2.imread('dhoni.png')
```

```python
# Creating New Image 
# Converting Original image to GreyScale image
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Inverting GreyScale Image
inverted_gray_img = 255 - gray_img
```

```python
# Applying Gaussian BLur to inverted image
blurred_img = cv2.GaussianBlur(inverted_gray_img,(21,21),0)
```

```python
# Inverting the blured image
inverted_blurred_img = 255 - blurred_img
```

```python
# Creating Pencil Sketch Image
pencil_sketch_img = cv2.divide(gray_img, inverted_blurred_img, scale=256.0)
```

```python
# Saving pencil sketch image with new name
cv2.imwrite("output_image.jpg",pencil_sketch_img)
```
